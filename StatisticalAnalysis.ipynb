```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('versicherung_x.csv', delimiter=';')
```

```python
data
```

```python
# Distribution of values
def normalizeKindAge(x):
    if x == "n.a.":
        x = 0
    return int(x)

def filterZero(x):
    return x != 0

child_data = data.copy()
child_data['aeltestesKind'] = child_data['aeltestesKind'].apply(normalizeKindAge)
child_data['juengstesKind'] = child_data['juengstesKind'].apply(normalizeKindAge)
child_data.filter(['juengstesKind', 'aeltestesKind']).hist()
plt.show()

# Without zero, because that makes the rest of the plot look tiny
child_data['aeltestesKind'] = child_data['aeltestesKind'].apply(normalizeKindAge)[filterZero]
child_data['juengstesKind'] = child_data['juengstesKind'].apply(normalizeKindAge)[filterZero]
child_data.filter(['juengstesKind', 'aeltestesKind']).hist()
plt.show()
```

```python
# Distribution of values
wage_data = data.copy()
wage_data['Familieneinkommen'].hist(bins=50)
```

```python
# Distribution of values
for col in data:
    if col in ['Familieneinkommen', 'aeltestesKind', 'juengstesKind']:
        continue
    age_data[col].value_counts().plot.bar()
    plt.show()
```
